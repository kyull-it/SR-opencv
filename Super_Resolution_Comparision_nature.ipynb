```python
# https://learnopencv.com/super-resolution-in-opencv/
import cv2
import matplotlib.pyplot as plt
```

```python
sample = Image.open(imagePath)
sample
```

```python
img = cv2.imread(imagePath)
img
# plt.imshow(img[:,:,::-1])
```

```python
img.shape
```

### FSRCNN_x2 모델 테스트

```python
startime = time.time()

# dnn super resolution_opencv
sr = cv2.dnn_superres.DnnSuperResImpl_create()

# .pb파일 path
path = "./cv2_dnn_sr/FSRCNN_x2.pb"
sr.readModel(path)

# set model
sr.setModel("fsrcnn", 2)          #3배 더 좋은 해상도 -> 숫자 변경가능
result1 = sr.upsample(img)

# ndarray to PIL image
pil_image1=Image.fromarray(result1[:,:,::-1])
# pil_image.show()
pil_image1.save('./test_img/result_fsrcnn_x2.jpeg')

endtime = time.time()
print("duration : ", endtime - startime)
```
