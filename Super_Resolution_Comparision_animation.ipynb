```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import time
```

```python
imagePath = "./test_img/DDoongE.jpg"
```

```python
sample = Image.open(imagePath)
sample
```

```python
img = cv2.imread(imagePath)
img.shape
```

### FSRCNN_x3

```python
startime = time.time()

# dnn super resolution_opencv
sr = cv2.dnn_superres.DnnSuperResImpl_create()

# .pb파일 path
path = "./cv2_dnn_sr/FSRCNN_x3.pb"
sr.readModel(path)

# set model
sr.setModel("fsrcnn",3)          #3배 더 좋은 해상도 -> 숫자 변경가능
result = sr.upsample(img)

# ndarray to PIL image
pil_image=Image.fromarray(result[:,:,::-1])
# pil_image.show()
pil_image.save('./test_img/DE_fsrcnn_x3.jpeg')

# Resized image
test_img = cv2.imread("./test_img/DE_fsrcnn_x3.jpeg")
test_resized = cv2.resize(test_img,dsize=None,fx=0.33,fy=0.33)

# ndarray to PIL image
test_pil_image=Image.fromarray(test_resized[:,:,::-1])
test_pil_image.save('./test_img/DE_resized_fsrcnn_x3.jpeg')

endtime = time.time()
print("duration : ", endtime - startime)
```
